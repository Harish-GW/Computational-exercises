```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def mandelbrot(c,n_max=300):
    z=0
    n=0
    while abs(z)<=2|n<=n_max:
        z=z*z+c
        n=n+1
    return n
```

```python
x=np.linspace(-1.5,1.5,2000)
y=np.linspace(-1.5,1.5,2000)

c=x+y[:,None]*1j
n=[[]*i for i in range(len(x))]
for i in range(len(x)):
    for j in range(len(y)):
        n[i].append(mandelbrot(c[i][j]))
```

```python
plt.pcolor(x,y,n)
plt.colorbar();
```
